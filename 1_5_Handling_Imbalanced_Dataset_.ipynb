# Imbalances Dataset:
A dataset with an unequal class distribution


```python
#importing the dependencies
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is Highly Imbalanced Dataset

0 --> Legit Transactions ,
1 --> Fraudulent Transactions

```python
#separating the legit and fradulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
Fradulent = credit_card_data[credit_card_data.Class == 1]
```

```python
legit.head()
```

```python
Fradulent.head()
```

```python
print(legit.shape, Fradulent.shape)
```

**UnderSampling The Data**

Build a sample dataset containig similar distribution of Legit & Fradulent Transactions

Here the number of fradulant transactions is 81

```python
legit_sample = legit.sample(n=81)
```

```python
legit_sample.shape
```

Now Concatenate the two data frame

```python
new_dataset = pd.concat([legit_sample , Fradulent] , axis= 0)
```
