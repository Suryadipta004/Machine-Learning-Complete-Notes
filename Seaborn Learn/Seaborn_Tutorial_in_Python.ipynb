**Line Plot**
*   create this using Seaborn


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x = [1,2,3,4,5,6,7,8,9]
y = [20,50,40,30,80,50,10,90,30]

df = pd.DataFrame({"Days":x , "No of People": y})
df.head(7)
```

```python
sns.lineplot(x="Days", y="No of People", data = df)
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.lineplot(x="day", y="total_bill", data = df , hue = 'sex', style = 'time' , palette='flare')
```

**Bar Plot**

```python
df = sns.load_dataset("titanic")
df.head(10)
```

```python
sns.barplot(x='who',y='fare',data= df ,hue='sex', color='red',saturation=0.85 , ci=False)
plt.show()
```

```python
sns.barplot(x='who',y='fare',data= df ,facecolor=[1,1,0,0],edgecolor="0.2")
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=df,order=["child" , "man" , "woman"])
plt.show()
```
