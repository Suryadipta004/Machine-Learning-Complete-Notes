#**Methods to Handle Missing Values:**
*   **Imputation**
*   **Dropping**

**Importing the libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset to a Pandas Dataframe
dataset = pd.read_csv('/Placement_Dataset.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

#**Central Tendencies:**
1.   **Mean**
2.   **Median**
3.   **Mode**

```python
# analyse the distribution of the data in the salary
sns.histplot(dataset['salary'], kde=True)
plt.show()
```

**Replace the missing values with Median Values**

```python
dataset.fillna(dataset['salary'].median() , inplace=True)
```

```python
dataset.isnull().sum()
```
