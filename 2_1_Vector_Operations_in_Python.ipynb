**Plotting a vector()**

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5 , scale_units='xy',angles='xy',scale=1)
plt.xlim(-1,8)
plt.ylim(-1,8)
plt.grid(True)
# plt.gca().set_aspect('equal')
plt.show()
```

```python
plt.quiver(0,0,4,5 , scale_units='xy',angles='xy',scale=1)
plt.quiver(0,0,-3,-6 , scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.grid(True)
# plt.gca().set_aspect('equal')
plt.show()
```

**Dot Product of two Vectors**

```python
c = np.array([40,20,50])
d = np.array([53,24,68])

c_dot_d = np.dot(c,d)
print(c_dot_d)
```

**Cross Product of two Vectors**

```python
c_cross_d = np.cross(c,d)
print(c_cross_d)
```

**Projection of vectors**

```python
print(np.sum(c**2))
```

```python
magnitude_of_d = np.sqrt(np.sum(d**2))
proj_of_c_on_d = (np.dot(c,d)/(magnitude_of_d)**2)*d
print(proj_of_c_on_d)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, c[0], c[1], c[2], color='r')
ax.quiver(0, 0, 0, d[0], d[1], d[2], color='b')
ax.quiver(0, 0, 0, proj_of_c_on_d[0], proj_of_c_on_d[1], proj_of_c_on_d[2], color='y')

ax.quiver(0, 0, 0, 10, 0, 0, color='black')  # X-axis
ax.quiver(0, 0, 0, 0, 10, 0, color='black')  # Y-axis
ax.quiver(0, 0, 0, 0, 0, 10, color='black')  # Z-axis

max_range = np.array([c, d, proj_of_c_on_d]).max() + 10
ax.set_xlim([0, max_range])
ax.set_ylim([0, max_range])
ax.set_zlim([0, max_range])


plt.show()
```
